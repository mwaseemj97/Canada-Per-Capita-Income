```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')

df.head(3)
```

```python
%matplotlib inline
plt.scatter(df.year,df.per_capita_income_usdollar, color='red', marker='+')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.per_capita_income_usdollar)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel("Year", fontsize=16)
plt.ylabel("Per_Capita_Income_US$", fontsize=16)
plt.scatter(df.year,df.per_capita_income_usdollar, color='red', marker = '+')
plt.plot(df.year, reg.predict(df[['year']]), color = 'blue')
```
